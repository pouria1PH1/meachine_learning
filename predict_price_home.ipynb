replace  true  and  false  with  numberic value 
replace name Address with random number for  calculating

```python
import pandas as pd
import numpy as np
import random

# Read the CSV file into a DataFrame
df = pd.read_csv(myfile)

# Map True values to [1] and False values to 0
df[[ 'Parking', 'Warehouse', 'Elevator']] = df[['Parking', 'Warehouse', 'Elevator']].applymap(lambda x: [0,1][int(bool(x))] if x in [True, False] else 0)

# Create a dictionary with random prices for each address
addresses = df['Address'].unique().tolist()
prices = [random.randint(1, 5) for _ in range(len(addresses))]
address_dict = dict(zip(addresses, prices))

# replace   new column of Address  with last column in the DataFrame with the random number for each address
df['Address'] = df['Address'].map(address_dict)


df


```

### cleaning data

```python

df["Area"].value_counts()
```

remove  row  that have  nothing values and not  suitable  data

```python
list_remove=[]
index=0
for  area in df["Area"]:
    if len(str(area))>3:
        list_remove+=[index]
    index+=1
    

list_remove

```

```python
df=df.drop(list_remove)
```

```python
# Drop rows with missing values
df=df.dropna()
df
```

split   test data and  train  data


```python
boolean=np.random.rand(len(df))<=0.8
train_data=df[boolean]
test_data=df[~boolean]
```

create  regression  model


```python
train_x=np.asanyarray(train_data[["Area","Room","Parking","Warehouse","Elevator"]])
train_y=np.asanyarray(train_data[["Price"]])
regr.fit(train_x,train_y)
print("coefficients:",regr.coef_)
print("intercept:",regr.intercept_)
```

predict  and check  model

```python
y_predicted=regr.predict(test_data[["Area","Room","Parking","Warehouse","Elevator"]])

real_test_y=np.asanyarray(test_data[["Price"]])



print("R2-score: %.2f" % r2_score(real_test_y,y_predicted) )
my_predict=regr.predict(([[60,2,1,1,1]]))
print(my_predict)
```
